### Define a function that differentiates between continuous and discrete variables.

```python
def split_continuous_discrete(df):
    
    discrete_columns = [col for col in df.columns if df[col].nunique() < 10]
    continuous_columns = [col for col in df.columns if df[col].nunique() > 10]

    return df[continuous_columns], df[discrete_columns]
```

```python
continuous_df, discrete_df = split_continuous_discrete(numerical_df)
```

#### Distribution with KDE

```python
def plot_continuous_df_dist(df):
    for col in df.columns:
        sns.distplot(df[col], kde=True)
        plt.title(col)
        plt.show()
```

```python
plot_continuous_df_dist(continuous_df)
```

#### Box plot

```python
def plot_continuous_df_box(df):
    for col in df.columns:
        sns.boxplot(x=df[col])
        plt.title(col)
        plt.show()
```

```python
plot_continuous_df_box(continuous_df)
```

#### Violin plot which combines both previous plots

```python
def plot_continuous_df_violin(df):
    for col in df.columns:
        sns.violinplot(x=df[col])
        plt.title(col)
        plt.show()
```

```python
plot_continuous_df_violin(continuous_df)
```

### There might be some columns that seem to be redundant, check their values to be sure. What should we do with them?

```python
#Val found this code
#for big chi2
import os as os
from itertools import product
import scipy.stats as ss

categorical_df.columns.values.tolist()
cat1 = ('state',
 'response',
 'coverage',
 'education',
 'employment_status',
 'gender',
 'location_code',
 'marital_status',
 'policy',
 'policy_type',
 'renew_offer_type',
 'sales_channel',
 'vehicle_class',
 'vehicle_size')
cat2 = ('state',
 'response',
 'coverage',
 'education',
 'employment_status',
 'gender',
 'location_code',
 'marital_status',
 'policy',
 'policy_type',
 'renew_offer_type',
 'sales_channel',
 'vehicle_class',
 'vehicle_size')
```

```python
# mix them into a big list to have them on both sides
cat_combos = list(product(cat1,cat2, repeat = 1))
```

```python
# Creating an empty variable and picking only the p value from the output of Chi-Square test
result = []
for i in cat_combos:
    if i[0] != i[1]:
        result.append((i[0],i[1],list(ss.chi2_contingency(pd.crosstab(
                            categorical_df[i[0]], categorical_df[i[1]])))[1]))
```

```python
#create a crosstab similar to corr()
chi_test_output = pd.DataFrame(result, columns = ['var1', 'var2','coeff'])
chi_test_output.round(4)

# Using pivot function to convert the above df into a crosstab
chi_test_output = chi_test_output.pivot(index='var1', columns='var2', values='coeff')

#finds values less than the set condition, which is our p_value
test = chi_test_output.lt(0.05)
test
```
